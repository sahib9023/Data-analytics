```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "Sample_Data_for_Activity.csv"  # Update the path if needed
df = pd.read_csv(file_path)

# Set Seaborn style
sns.set(style="whitegrid")

# Plot the distribution of the Normal_Distribution column
sns.displot(df["Normal_Distribution"], kde=True, bins=30, aspect=1.5)

# Set title
plt.title("Distribution of Normal_Distribution")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Jointplot: Scatter + KDE for Normal vs Uniform Distribution
sns.jointplot(x=df["Normal_Distribution"], y=df["Uniform_Distribution"], kind="scatter")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot: Visualize relationships among all numerical columns
sns.pairplot(df)

# Show plot
plt.show()
```
